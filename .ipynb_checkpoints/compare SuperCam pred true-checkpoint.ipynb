```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score
from math import sqrt
```

```python
df = pd.read_csv('P:\\SUPERCAM_from_PDS\\LIBS_RDR_metadata_w_pred_comps.csv')
df = df[df.target.str.contains('scct')].sort_values('target', ignore_index=True)
```

```python
comp_cols = [c for c in df.columns if 'O' in c and 'stdev' not in c]
mean_cols = [c+'_mean' for c in comp_cols]
median_cols = [c+'_median' for c in comp_cols]
stdev_cols = [c for c in df.columns if 'stdev' in c]
stdev_mean_cols = [c+'_mean_listed_stdev' for c in comp_cols]

for l in [comp_cols,mean_cols,stdev_cols,stdev_mean_cols,median_cols]:
    l.insert(0,'target')
```

```python
# number of duplicates
dup_df = pd.DataFrame(df.groupby('target').size()).reset_index().rename(columns={0:'n_dups'})

# average value
mean_df = df[comp_cols].groupby('target').mean().reset_index()
mean_df.columns = mean_cols

# median value
med_df = df[comp_cols].groupby('target').median().reset_index()
med_df.columns = median_cols

# stdev of values
stdev_df = df[comp_cols].groupby('target').std().reset_index()
stdev_df.columns = stdev_cols

# mean stdev provided for values
mean_stdev_df = df[stdev_cols].groupby('target').mean().reset_index()
mean_stdev_df.columns = stdev_mean_cols

# merge results
results_df = dup_df.merge(mean_df).merge(med_df).merge(stdev_df).merge(mean_stdev_df)
```

```python
# actual values from calibration metadata
actual_df = pd.read_csv('G:\\My Drive\\Darby Work\\SuperCam calibration\\data\\SuperCam_cal_shift-2pix_metadata.csv')
actual_cols = list(comp_cols[1:])
actual_cols.insert(0,'Target_Name')
actual_df = actual_df[actual_cols].drop_duplicates()
actual_cols = [a+'_actual' for a in actual_cols[1:]]
actual_cols.insert(0,'Target_Name')
actual_df.columns = actual_cols

# key to match meta target name to SCCT target name
sample_key = pd.read_excel('P:\\SUPERCAM_from_PDS\\SCCT_key.xlsx', skiprows=2)
sample_key = sample_key[['Target Name','Metadata Name']]

# add key
results_df = results_df.merge(sample_key, how='left', left_on='target', right_on='Target Name').drop(columns='Target Name')
# add actual compositions
results_df = results_df.merge(actual_df, how='left', left_on='Metadata Name', right_on='Target_Name').drop(columns=['Metadata Name', 'Target_Name'])
```

```python
# sort columns
all_cols = list(results_df.columns[2:])
all_cols.sort()
all_cols.insert(0,'target')
all_cols.insert(1,'n_dups')
results_df = results_df[all_cols]
```

```python
results_df.to_csv('results_summary.csv', index=False)
```

### get RMSEs

```python
# get reported accuracy
reported_stats = pd.read_csv('P:\\SUPERCAM_from_PDS\\supercam_libs_moc.csv', nrows=7)
reported_stats = reported_stats[[c for c in reported_stats.columns if 'Unnamed' not in c]]

stats_dict = dict()

for oxide in comp_cols[1:]:
    
    report_rmse = reported_stats[reported_stats['Training set Quartiles']=='RMSEP'][oxide].values[0]

    true = oxide+'_actual'
    pred_mean = oxide+'_mean'
    pred_median = oxide+'_median'
    
    # remove those with NA if there
    temp = results_df.dropna(subset=[true,pred_mean])
    # get data
    true = temp[true].values
    pred_mean = temp[pred_mean].values
    pred_median = temp[pred_median].values
    
    rmse_mean = sqrt(mean_squared_error(true, pred_mean))
    r2_mean = r2_score(true, pred_mean)
    rmse_median = sqrt(mean_squared_error(true, pred_median))
    r2_median = r2_score(true, pred_median)
    
    stats_dict[oxide] = {
        'reported_rmsep':report_rmse,
        'rmse_mean':round(rmse_mean,2),
        'rmse_median':round(rmse_median,2),
        'r2_mean':round(r2_mean,2),
        'r2_median':round(r2_median,2)
    }
```

```python
# make predicted vs true figures
for oxide in comp_cols[1:]:
    
    stats = stats_dict[oxide]
    
    mean_c = 'navy'
    med_c = 'orange'
    
    true = oxide+'_actual'
    pred_mean = oxide+'_mean'
    pred_median = oxide+'_median'
    error = oxide+'_stdev'
#    title = f'''{oxide} (wt%)    Reported RMSEP: {stats['reported_rmsep']}
#    Mean value R2: {stats['r2_mean']}    Mean value RMSE: {stats['rmse_mean']}
#    Median value R2: {stats['r2_median']}    Mean value RMSE: {stats['rmse_median']}'''
    title = f"{oxide} (wt%)\n    R2: {stats['r2_mean']}    RMSE: {stats['rmse_mean']}    Reported RMSEP: {stats['reported_rmsep']}"
    
    # remove those with NA if there
    temp = results_df.dropna(subset=[true,pred_mean])
    
    # get data
    x = temp[true].values
    y_mean = temp[pred_mean].values
    yerr = temp[error].values
    y_median = temp[pred_median].values
    
    plt.figure(figsize=(6,6))
    
    # mean
    plt.scatter(x,y_mean,alpha=0.6,c=mean_c,label='Mean')
    plt.errorbar(x,y_mean, yerr=yerr, linestyle='', c=mean_c)
    
    # median
#    plt.scatter(x,y_median,alpha=0.7,c=med_c,label='Median')
    
    # 1:1 line
    xmin, xmax = plt.xlim()
    ymin, ymax = plt.ylim()
    plt_min = min(ymin, xmin)
    plt_max = max(ymax, xmax)

    plt.plot([plt_min,plt_max], [plt_min,plt_max], 'k--', linewidth=1)
    plt.xlim(xmin,xmax)
    plt.ylim(ymin,ymax)
    
    plt.xlabel('True')
    plt.ylabel('Predicted')
    plt.title(title)
    plt.tight_layout()
    plt.savefig(f'SCCT pred true plots\\{oxide}.jpg', dpi=600)
    plt.savefig(f'SCCT pred true plots\\{oxide}.eps', dpi=600)
    plt.show()
```
